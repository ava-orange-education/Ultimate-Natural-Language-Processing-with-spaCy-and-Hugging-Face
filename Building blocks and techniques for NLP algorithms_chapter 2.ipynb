Bag of Words model

```python
import nltk
# Define your documents
documents = ["I am a good man", "You are a good man too"]

# Function to tokenize documents (using split())
def tokenize_documents(documents):
    all_tokens = []
    for document in documents:
        all_tokens.extend(document.split())  # Extend list with split tokens
    return all_tokens

# Get all the unique tokens
all_tokens = list(set(tokenize_documents(documents)))

# Print all tokens
print("All Tokens:", all_tokens)
```

```python
# Function to create bag-of-words model
def create_bow_model(tokens):
    vocabulary = set(tokens)  # Create vocabulary (unique words)
    bow_model = {word: 0 for word in vocabulary}  # Initialize dictionary
    for token in tokens:
        bow_model[token] += 1  # Count word occurrences
    return bow_model

# Get all tokens
all_tokens = tokenize_documents(documents)

# Create bag-of-words model
bow_model = create_bow_model(all_tokens)

# Print bag-of-words model (dictionary)
print("Bag-of-Words Model:", bow_model)
```

N-grams

```python
from nltk.tokenize import word_tokenize

# Sample text
text = "Natural Language Processing is a powerful tool for analyzing text data."

# Function to generate n-grams
def generate_n_grams(text, n):
    tokens = word_tokenize(text.lower())  # Tokenize and lowercase text
    n_grams = []
    for i in range(len(tokens)-n+1):
        n_gram = " ".join(tokens[i:i+n])  # Join tokens with space
        n_grams.append(n_gram)
    return n_grams

# Generate bigrams and trigrams
bigrams = generate_n_grams(text, 2)
trigrams = generate_n_grams(text, 3)

# Print n-gram lists
print("Bigrams:", bigrams)
print("Trigrams:", trigrams)
```

```python
from nltk.util import ngrams

# Example sentence
sentence = "Natural Language Processing is a powerful tool for analyzing text data."

# Tokenize the sentence into words
words = sentence.split()

# Create bigrams from the list of words
bigrams = ngrams(words, 4)

# Print the bigrams
for bigram in bigrams:
    print(bigram)
```

TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample documents
documents = ["I am a good man.",
              "You are a good man too."]

# Create TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit the vectorizer on the documents
tfidf_matrix = vectorizer.fit_transform(documents)

# Print TF-IDF matrix (sparse matrix)
print(tfidf_matrix.toarray())

# Print feature names (vocabulary)
print(vectorizer.get_feature_names_out())
```
